# Processing script to run multiple folders with the same calibration file

```python
from multiplane import MultiplaneProcess
import os
```

## set parameters

```python
save_individual = False      # True for SMLM, False for SOFI, QPI
save_in_subfolders = False  # save each plane in a separate subfolder
i_corr = True              # apply intensity correction, False for SMLM, True for QPI
apply_transform = True     # apply the affine transform before saving data, True for QPI and SOFI, False for SMLM 
dF_batch = 1
```

## select processing folders 

```python
data_path = r'F:\moritz\Technical_paper\4x2_scoptonics\20260422_livecell\data\FOV2'
```

## select calibration folder 

```python
cal_dummy = MultiplaneProcess()
cal_dummy.path = r'F:\moritz\Technical_paper\4x2_scoptonics\20260422_livecell\beadstack\PS_e561nm_dz20nm_1\cal_data' #data_path
cal_dummy.load_calibration()
```

## start processing

```python
for folder in os.listdir(data_path):
    proc = MultiplaneProcess()
    proc.select_data_directory(os.path.join(data_path, folder))
    proc.cal = cal_dummy.cal

    proc.save_individual = save_individual
    proc.save_in_subfolders = save_in_subfolders
    proc.i_corr = i_corr
    proc.P['apply_transform'] = apply_transform
    proc.P['dF_batch'] = dF_batch

    proc.execute()

print('Batch processing complete.')
```
